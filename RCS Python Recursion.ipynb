## Building a recursive program

```python
def factorial(n):
    if n <= 1: #Base Case very important!! Start here
        return 1
    return n*fract(n-1)
```

```python
factorial(100)
```

```python
## Stack frame is a limited resourc
```

```python
factorial(5010)
```

```python
# A fix of sorts would be to increase stack frame limit with
import sys
sys.setrecursionlimit(5200)
```

```python
factorial(1470)
```

Some tutorials put base case in else block, but that obscures the meaning of base case.

<h2>**Base case should be decided on first!**</h2>

```python
# For fractals, regular loop is actually better(doesnt use up stack memory as much)
def factloop(n):
    acc = 1
    for i in range(n,1,-1):
        acc*=i
    return acc
```

```python
factloop(10)
```

```python
factloop(5010)
```

```python
## All recursive programs can be rewritten as normal loop just but they can appear longer
```

```python
%timeit factorial(1000)
```

```python
%timeit factloop(1000)
```

## Recursion can solve some problems very elegantly

Many code challenges can be solved quicker with recursion - need to figure out a base case then work on reducing to that base case

Rarely used in production because of stack overflow(running out of stack frames), 
tail call optimization problems(regular Python does not support tail call optimzation)

Each Stack Frame can potentially use a lot of memory if using many local variables
